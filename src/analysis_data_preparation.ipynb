<p>The null values in columns such as winner_entry, winner_seed, loser_entry and loser_seed,are not important at all</p>

<h2>let's impute other columns that has null values that are important such as the winner_rank and loser_rank </h2>

```python
#checking the distribution plot of the winner rank to decided with imputation method to use
sns.displot(df_tennis['winner_rank'])
```

```python
df_tennis['winner_rank'].fillna(df_tennis['winner_rank'].median(), inplace=True)
df_tennis['loser_rank'].fillna(df_tennis['loser_rank'].median(), inplace=True)
```

<h2>let's impute other columns that has null values that are important such as the winner_ht (winner height) and loser_ht(loser height) </h2>

<p>This is column is going to take into account for the model </p>

```python
sns.boxplot(x=df_tennis['winner_ht'])
```

```python
#we are going to fill the null values with the mean because the distribution is not skewed

df_tennis['winner_ht'].fillna(df_tennis['winner_ht'].mean(), inplace=True)
df_tennis['loser_ht'].fillna(df_tennis['loser_ht'].mean(), inplace=True)
```

<h2>let's impute other columns that has null values that are important such as the winner_age and loser_age </h2>

<p>This is column is going to take into account for the model </p>

```python
sns.boxplot(x=df_tennis['winner_age'])
```

```python
sns.boxplot(x=df_tennis['loser_age'])
```

```python
#we are going to fill the null values with the meadian because the distribution is skewed
df_tennis['winner_age'].fillna(df_tennis['winner_age'].median(), inplace=True)
df_tennis['loser_age'].fillna(df_tennis['loser_age'].median(), inplace=True)
```

<h2> Let's drop some columns that won't contribute to the model </h2>

```python
columns_to_drop=['winner_entry', 'loser_entry', 'winner_seed', 'loser_seed', 'winner_ioc', 
                    'loser_ioc', 'winner_rank_points', 'loser_rank_points']

#drop the columns
df_tennis.drop(columns_to_drop, axis=1, inplace=True)
```

<p>It appears that the null values are associated with the Davis Cup. Upon analysis, we recognize that the Davis Cup is a team competition played by countries. Since our model is designed to predict individual player matchups, we will exclude rows related to the Davis Cup from our dataset.</p>


```python
pattern="Davis Cup"
filter=df_tennis['tourney_name'].str.contains(pattern)
df_tennis=df_tennis[~filter]
```

<p>Let's fill the stats value such as w_ace, w_df, among others with zeros </p>

```python
fill_values={
    'w_ace':0,
    'w_df':0,
    'w_svpt':0,
    'w_1stIn':0,
    'w_1stWon':0,
    'w_2ndWon':0,
    'w_SvGms':0,
    'w_bpSaved':0,
    'w_bpFaced':0,
    'l_ace':0,
    'l_df':0,
    'l_svpt':0,
    'l_1stIn':0,
    'l_1stWon':0,
    'l_2ndWon':0,
    'l_SvGms':0,
    'l_bpSaved':0,
    'l_bpFaced':0,
    

}

#let's apply the dictionary using the fillna method
df_tennis.fillna(value=fill_values, inplace=True)

#check missing values again
df_tennis.isna().sum()
```

<p>still having missing values in the score and minutes column
as the score only have two missing values we can drop them</p>

```python
df_tennis.dropna(subset=['score'], inplace=True)
```

<p>Let's work now on the minutes column</p>

```python

sns.displot(df_tennis['minutes'])
```

```python
df_tennis['minutes'].fillna(df_tennis['minutes'].mean(), inplace=True)

#checking missing values again
df_tennis.isna().sum()
```

<p>Perfect Not missing values </p>

<h2>We are to organize the dataset by player and not by matche </h2>

<p>We are also to assign a target column. If the player won the match is 1, otherwise zero </p>

```python
df_winners=df_tennis[['winner_id', 'winner_name', 'winner_hand', 'winner_ht', 'winner_age',
                'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms', 'w_bpSaved', 'w_bpFaced', 'winner_rank', 'loser_rank', 'surface', 'tourney_level', 'tourney_date', 'best_of']].copy()
df_winners['target']=1

df_losers=df_tennis[['loser_id', 'loser_name', 'loser_hand', 'loser_ht', 'loser_age',
                'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced', 'loser_rank', 'winner_rank', 'surface', 'tourney_level', 'tourney_date', 'best_of']].copy()
df_losers['target']=0
```

```python
df_winners.columns=['player_id', 'player_name', 'player_hand', 'player_ht', 'player_age', 
                    'ace', 'df', 'svpt', '1stIn', '1stWon', '2ndWon', 'SvGms', 'bpSaved', 'bpFaced', 'rank', 'opponent_rank', 'surface', 'tourney_level', 'tourney_date', 'best_of', 'target']
df_losers.columns=['player_id', 'player_name', 'player_hand', 'player_ht', 'player_age',
                    'ace', 'df', 'svpt', '1stIn', '1stWon', '2ndWon', 'SvGms', 'bpSaved', 'bpFaced', 'rank', 'opponent_rank', 'surface', 'tourney_level', 'tourney_date', 'best_of', 'target']

df_players=pd.concat([df_winners, df_losers], ignore_index=True)
```

```python
#sort the players by player id
df_players_sorted=df_players.sort_values('player_id')
df_players_sorted.head()
```

<h2> Let's extract information from the tourney_date column </h2>

<p>We are going to convert this column to show only month </h2>

```python
df_players_sorted['tourney_date']=pd.to_datetime(df_players_sorted['tourney_date'],format='%Y%m%d')
df_players_sorted['month']=df_players_sorted['tourney_date'].dt.month #extracting month from the date
```

```python
#checki null values
df_players_sorted.isna().sum()
```

```python
#check data types
df_players_sorted.dtypes
```

```python
#we have float values, let's convert them to int
df_players_sorted['player_ht']=df_players_sorted['player_ht'].astype(int)
df_players_sorted['player_age']=df_players_sorted['player_age'].astype(int)
df_players_sorted['ace']=df_players_sorted['ace'].astype(int)
df_players_sorted['df']=df_players_sorted['df'].astype(int)
df_players_sorted['svpt']=df_players_sorted['svpt'].astype(int)
df_players_sorted['1stIn']=df_players_sorted['1stIn'].astype(int)
df_players_sorted['1stWon']=df_players_sorted['1stWon'].astype(int)
df_players_sorted['2ndWon']=df_players_sorted['2ndWon'].astype(int)
df_players_sorted['SvGms']=df_players_sorted['SvGms'].astype(int)
df_players_sorted['bpSaved']=df_players_sorted['bpSaved'].astype(int)
df_players_sorted['bpFaced']=df_players_sorted['bpFaced'].astype(int)
df_players_sorted['rank']=df_players_sorted['rank'].astype(int)
df_players_sorted['opponent_rank']=df_players_sorted['opponent_rank'].astype(int)
```

```python
df_players_sorted.drop(['tourney_date'], axis=1, inplace=True)
```

```python
df_players_sorted.head()
```

```python
df_players_sorted.to_csv('../data/prepared_data.csv', index=False)
```

<h2> Let's create a different dataset with stats per player, surface, and toruney level </h2>

```python
players_stats_average=df_players_sorted.groupby(['player_name', 'surface', 'tourney_level']).agg({
    'ace':'mean',
    'df':'mean',
    'svpt':'mean',
    '1stIn':'mean',
    '1stWon':'mean',
    '2ndWon':'mean',
    'SvGms':'mean',
    'bpSaved':'mean',
    'bpFaced':'mean',    
}).reset_index()

players_stats_average.columns=['player_name', 'surface', 'tourney_level', 'avg_ace', 'avg_df', 'avg_svpt', 'avg_1stIn', 'avg_1stWon', 'avg_2ndWon', 'avg_SvGms', 'avg_bpSaved', 'avg_bpFaced']

players_stats_average.head(50)
```

```python
players_stats_average.to_csv('../data/players_stats_average.csv', index=False)
```
